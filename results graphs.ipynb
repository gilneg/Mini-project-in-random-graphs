```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([100000, 250000,500000, 750000, 1000000])

y1 = np.array([2118.8, 3023.8, 5289.4, 7534.2, 8652.7])
# plotting the line 1 points
plt.plot(x1, y1, marker='o', label = "L1")

# line 2 points
x2 = np.array([100000, 250000,500000, 750000, 1000000])

y2 = np.array([803, 1536.8, 2217.6, 3092, 3601.5])
# plotting the line 2 points
plt.plot(x2, y2, marker='o', label = "L2")

#plotting ln(n)
plt.plot(x2, 2.1*np.log(x2), marker='o', label = "2.1*ln(n) ϵ Θ(ln(n))")

# naming the x axis
plt.xlabel('number of vertices')
# naming the y axis
plt.ylabel('size of the component')
# giving a title to my graph
plt.title('Very Subcritical')

# show a legend on the plot
plt.legend()

# function to show the plot
plt.show()
```
